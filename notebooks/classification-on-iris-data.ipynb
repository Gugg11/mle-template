# Классификация семян пшеницы

В данной работе используется датасет seeds.csv.  
Цель задачи — классифицировать тип семян пшеницы по числовым признакам:

- Area
- Perimeter
- Compactness
- Kernel.Length
- Kernel.Width
- Asymmetry.Coeff
- Kernel.Groove

Целевая переменная — Type.


**Part I: Import library and load data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

df = pd.read_csv("../data/seeds.csv")
```

**Part II: Check the data information**

```python
# first five observations
df.head()
```

**Part III: Explorary data analysis**

```python
df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
df["Type"].value_counts().plot(kind="bar")
plt.title("Распределение классов Type")
plt.xlabel("Класс")
plt.ylabel("Количество объектов")
plt.show()
```

**Part 4: Train Test Split**

```python
X = df.drop(columns=["Type"])
y = df["Type"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=0,
    stratify=y
)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

**Part 5: Scaling**



```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Part 5: Train a Model**

В качестве базовой модели используется Logistic Regression.  
Это классический алгоритм классификации, подходящий для табличных данных.

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

**Part 6: Model Evaluation**

```python
pred = model.predict(X_test_scaled)

print("Confusion matrix:")
print(confusion_matrix(y_test, pred))

print("\nClassification report:")
print(classification_report(y_test, pred))

print("\nAccuracy score:", accuracy_score(y_test, pred))
```
